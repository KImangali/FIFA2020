    Introduction

    FIFA 20 is a football simulation video game published by Electronic Arts as part of the FIFA series. Each player in FIFA has an overall rating as well as six scores for the key stats; Pace, Shooting, Passing, Dribbling, Defending, and Physical. These stats are combined with a player's international recognition to calculate the player's overall rating.

short_name - name of the player<br>
age - age of the player<br>
nationality - country of the player<br>
club - his football club where he plays<br>
overall - player rating(0-99) in the game<br>
value_eur - player value in euro<br>
wage_eur - player salary in euro<br>
preferred_foot - leading leg<br>
international_reputation - player's recognition internationally, player popularity<br>
team_position - player's position in the club<br>

##  1. WEB scrapping and data manipulation

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
```

```python
data = [] #list for saving scrapping data
num_pages = 9 #500 players in 9 pages
pages_url = []  #for add urls of 9 pages
count_player = 0 #count of player that offseting on the select

for x in range(1, num_pages + 1):
    tail = '&offset=' + str(count_player) #tail of the link
    link = 'https://sofifa.com/players?type=all&oah=99&hl=en-US&showCol%5B0%5D=ae&showCol%5B1%5D=pf&showCol%5B2%5D=oa&showCol%5B3%5D=pt&showCol%5B4%5D=gu&showCol%5B5%5D=vl&showCol%5B6%5D=ir&showCol%5B7%5D=pac&showCol%5B8%5D=sho&showCol%5B9%5D=pas&showCol%5B10%5D=dri&showCol%5B11%5D=def&showCol%5B12%5D=phy&showCol%5B13%5D=wg' + str(tail)
        #adding tail to the end of link
    
    pages_url.append( requests.get(link) )
    count_player += 60
print('Competed')
```

```python
for r in pages_url:   
    html = BS(r.content, 'lxml')
    
    i = 0
    players = 60
    for element in html.select('tbody'):
        name = element.select('tr > td.col-name > a.tooltip > .bp3-text-overflow-ellipsis')
        age = element.select('tr > td.col-ae')
        country = element.select('tr > td.col-name > a.tooltip > .bp3-text-overflow-ellipsis > img')
        rating = element.select('tr > td.col-oa > span')
        club = element.select('tr > td.col-name > .bp3-text-overflow-ellipsis > a')
        foot = element.select('tr > td.col-pf')
        growth = element.select('tr > td.col-gu > span')
        value = element.select('tr > td.col-vl')
        wage = element.select('tr > td.col-wg')
        reputation = element.select('tr > td.col-ir')

        while i < players:

            n = name[i].text
            ae = age[i].text
            cntr = country[i].get('title')
            rtg = rating[i].text
            clb = club[i].text
            ft = foot[i].text
            grwth = growth[i].text
            vl = value[i].text
            wg = wage[i].text
            rpttn = reputation[i].text

            data.append([n, ae, cntr, rtg, clb, ft, grwth, vl, wg, rpttn])
            i += 1
print('Completed')
```

```python
print(len(data))
data = data[0:500]
print(len(data))
```

```python
fifa2020_data = pd.DataFrame(data, columns=['name', 'age', 'country', 'rating', 'club', 'foot', 'growth', 'value', 'wage', 'reputation'])
fifa2020_data.head(5)
```
